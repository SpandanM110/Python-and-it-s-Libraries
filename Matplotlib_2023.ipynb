```python
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
# np.random.seed(19680801)


N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
# Equation of line
m=10
b=5
x=np.array([1,8,23,34])
y=m*x+b
plt.plot(x,y)
plt.show()
```

STYLES IN MATPLOTLIB

```python
N=50
x=np.array([1,2,3,4])
y=np.array([23,12,45,32])

plt.plot(x,y,color='orange',marker='o')
plt.show()


plt.plot(x,y,color='orange')
plt.show()

plt.plot(x,y,"bo")
plt.show()

plt.plot(x,y,"b+")
plt.show()

plt.plot(x,y,"b--")
plt.show()

plt.plot(x,y,"k^:")
plt.show()

plt.plot(x,y,"b--",linewidth=5)
plt.show()

```

```python
x=np.array([1,2,3,4,5])
y=np.array([2,6,76,32,21])
plt.plot(x,y)
plt.ylabel("Random values")
plt.xlabel("Pointers")
plt.title("Plotting a randomn graph")
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2,6,76,32,21])
plt.plot(x,y)

#to plot at your own defined axis
plt.axis([0,6,0,20])
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2,6,76,32,21])
plt.plot(x,y)

plt.grid(True)
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2,6,76,32,21])
plt.xscale('linear')
plt.yscale('log')
plt.plot(x,y)
plt.show()
```

```python
import explode
labels='India','US','Uk','Germany'
sizes=[38,32,20,10]
explode=(0.1,0.2,0,0)

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

```python
mu=100
sigma=15
x=mu+sigma*np.random.randn(10000)

n,bins,patches=plt.hist(x,50,normed=1, facecolor='blue', alpha=0.80)

plt.xlabel('Smart graphs')
plt.ylabel('Probabilitis')
plt.axis([40,160,0,0.03])
plt.grid(True)

plt.show()
```
